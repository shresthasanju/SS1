```python
#program for gaussian plot (1st program)
import numpy as np
xs = np.arange(1,6,0.001)
xs
sum(xs)
mxs=np.mean(xs)
#xs = np.arange(1,6,1)
#xs---------------------(2nd porgram)
sxs=sum(xs)
mxs=np.mean(xs)
sdxs=np.std(xs)
print('xs=',xs,'sxs=',sxs,'mxs=',mxs,'sdxs=',sdxs)
#---------(3rd program)
#N = 1/np.sqrt(2*np.pi*sigma**2)
gau=np.exp(-(xs-mxs)**2/(2*sdxs**2))/np.sqrt(2*np.pi*sdxs**2)
gau
#-----------------------(4th program)
import pandas as pd
DF = pd.DataFrame(gau,xs)
DF
#____________(5th program)
import matplotlib.pyplot as plt
plt.figure(figsize = [10,2])
plt.plot(xs,gau)
```
